```python
import requests
import json
from pprint import pprint

park_url = "https://tcgbusfs.blob.core.windows.net/blobtcmsv/TCMSV_alldesc.json"
remain_url = "https://tcgbusfs.blob.core.windows.net/blobtcmsv/TCMSV_allavailable.json"


HEADERS = {
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_6) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/86.0.4240.80 Safari/537.36"
}

park_respond = requests.get(park_url, headers=HEADERS)
park_list = park_respond.json()["data"]["park"]

remain_respond = requests.get(remain_url, headers=HEADERS)
remain_list = remain_respond.json()["data"]["park"]
pprint(park_list[0])
pprint(remain_list[0])
```
